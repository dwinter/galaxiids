Parse our the interesting lines from stdout

```python
def has_digit(s):
    """ """
    for char in s:
        if char.isdigit():
            return True
    return False

def parse_line(line):
    """ """ 
    try:
        contig, rest = line.split(":")
    except ValueError: #end of file
        return None
    if not rest.startswith("\t No microsatellites found"):
        return [contig] + [s for s in rest.split() if has_digit(s)]
        
def filter_msat_file(fname, outname):
    """ """
    line_count = 0
    with open(fname) as msats:
        with open(outname, "w") as out:
            out.write("contig, motif, start, stop\n")
            for line in msats:
                parsed = parse_line(line)
                if parsed:
                    line_count += 1
                    out.write(",".join(parsed) + "\n")
    print "wrote data for microsattelites from {0} contigs".format(line_count)


filter_msat_file("dep_msats.tsv", "filtered_dep_msats.csv")
filter_msat_file("brev_msats.tsv", "filtered_brev_msats.csv")

                    
                
```
